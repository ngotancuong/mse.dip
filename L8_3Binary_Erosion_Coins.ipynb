```python
import matplotlib.pyplot as plt
import cv2
import urllib.request
import numpy as np


# Load image
url = 'https://raw.githubusercontent.com/ngotancuong/mse.dip/main/images/circles.png'
with urllib.request.urlopen(url) as url:
    s = url.read()
    img = cv2.imdecode(np.asarray(bytearray(s), dtype=np.uint8), cv2.IMREAD_UNCHANGED)

# Perform erosion with square with edge = 30
se1 = cv2.getStructuringElement(cv2.MORPH_RECT, (30, 30))
bw1 = cv2.erode(img, se1)

# Perform erosion with disk with radius = 30
se2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (30, 30))
bw2 = cv2.erode(img, se2)

# Perform erosion with rectangle with dimensions = [15 30]
se3 = cv2.getStructuringElement(cv2.MORPH_RECT, (30, 15))
bw3 = cv2.erode(img, se3)

# Perform erosion with diamond with radius = 30
se4 = cv2.getStructuringElement(cv2.MORPH_CROSS, (30, 30))
bw4 = cv2.erode(img, se4)

# Perform erosion with line with length = 30 and degree = 45
se5 = cv2.getStructuringElement(cv2.MORPH_RECT, (30, 1))
se5 = cv2.warpAffine(se5, cv2.getRotationMatrix2D((15, 0), 45, 1.0), (30, 30))
bw5 = cv2.erode(img, se5)

# Show and save images
titles = ['Original Image', 'Eroded by Square', 'Eroded by Disk', 'Eroded by Rectangle', 'Eroded by Diamond', 'Eroded by Line']
images = [img, bw1, bw2, bw3, bw4, bw5]
fig, axs = plt.subplots(2, 3)
for i, ax in enumerate(axs.flat):
    ax.imshow(images[i], cmap='gray')
    ax.set_title(titles[i])
plt.show()

cv2.imwrite('Binary_Erosion_Coins_Square.jpg', bw1)
cv2.imwrite('Binary_Erosion_Coins_Disk.jpg', bw2)
cv2.imwrite('Binary_Erosion_Coins_Rectangle.jpg', bw3)
cv2.imwrite('Binary_Erosion_Coins_Diamond.jpg', bw4)
cv2.imwrite('Binary_Erosion_Coins_Line.jpg', bw5)
```
