```python
import matplotlib.pyplot as plt
import cv2
import urllib.request
import numpy as np

# Load image from URL
url = 'https://raw.githubusercontent.com/ngotancuong/mse.dip/main/images/noisy_image.jfif'
with urllib.request.urlopen(url) as url_response:
    img_array = np.asarray(bytearray(url_response.read()), dtype=np.uint8)
    Img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

# Apply two different 3x3 sharpening filters
h1 = np.array([[0, -1, 0], [-1, 8, -1], [0, -1, 0]]) / 4
h2 = np.array([[0, -1, 0], [-1, 7, -1], [0, -1, 0]])
filteredImg1 = cv2.filter2D(Img, -1, h1, borderType=cv2.BORDER_REPLICATE)
filteredImg2 = cv2.filter2D(Img, -1, h2, borderType=cv2.BORDER_REPLICATE)

# Show images
plt.figure(figsize=(20,20))
plt.subplot(2, 2, 1), plt.imshow(cv2.cvtColor(Img, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(2, 2, 2), plt.imshow(cv2.cvtColor(filteredImg1, cv2.COLOR_BGR2RGB)), plt.title('Sharpening Filtered Image 1')
plt.subplot(2, 2, 3), plt.imshow(cv2.cvtColor(Img, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(2, 2, 4), plt.imshow(cv2.cvtColor(filteredImg2, cv2.COLOR_BGR2RGB)), plt.title('Sharpening Filtered Image 2')
plt.show()

# Save images
cv2.imwrite('Sharpening_Filter1.png', filteredImg1)
cv2.imwrite('Sharpening_Filter2.png', filteredImg2)
```
