```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request

#load image
tree = urllib.request.urlopen('https://raw.githubusercontent.com/ngotancuong/mse.dip/main/images/tree.jfif')
arr = np.asarray(bytearray(tree.read()), dtype=np.uint8)
Img = cv2.imdecode(arr, -1) # 'Load it as it is'#
#Img = cv2.imread('tree.jfif', cv2.IMREAD_GRAYSCALE)
# convert to grayscale
Img = cv2.cvtColor(Img, cv2.COLOR_BGR2GRAY) 

#calculate histogram
hist = cv2.calcHist([Img], [0], None, [256], [0, 256])

#plot histogram
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
ax1.bar(np.arange(256), hist.ravel())
ax1.set_xlim([0, 255])
ax1.set_ylim([0, 5000])
ax1.set_xlabel('Intensity')
ax1.set_ylabel('# of pixels')
ax1.set_title('Histogram')

#show and save image
ax2.imshow(cv2.cvtColor(Img, cv2.COLOR_BGR2RGB))
ax2.set_title('Original Image')
plt.savefig('Histogram.png', dpi=300)
plt.show()
```
