```python
import urllib.request
import cv2
import numpy as np
import matplotlib.pyplot as plt

# load image
url = 'https://raw.githubusercontent.com/ngotancuong/mse.dip/main/images/noisy_image.jfif'
req = urllib.request.urlopen(url)
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
Img = cv2.imdecode(arr, -1) / 255.0

# construct filter impulse responses
h1 = np.ones((1,5))/5 #horizontal filter
h2 = np.ones((5,1))/5 #vertical filter
h3 = np.ones((5,5))/25 #box/window filter

# perform filtering on the image
filteredImg1 = cv2.filter2D(Img, -1, h1, borderType=cv2.BORDER_REFLECT)
filteredImg2 = cv2.filter2D(Img, -1, h2, borderType=cv2.BORDER_REFLECT)
filteredImg3 = cv2.filter2D(Img, -1, h3, borderType=cv2.BORDER_REFLECT)

# show images
plt.rcParams['figure.figsize'] = [20, 8]
plt.figure(1)
plt.subplot(1, 2, 1)
plt.imshow(Img)
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(filteredImg1)
plt.title('Horizontal Filtered Image')

plt.rcParams['figure.figsize'] = [20, 8]
plt.figure(2)
plt.subplot(1, 2, 1)
plt.imshow(Img)
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(filteredImg2)
plt.title('Vertical Filtered Image')

plt.rcParams['figure.figsize'] = [20, 8]
plt.figure(3)
plt.subplot(1, 2, 1)
plt.imshow(Img)
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(filteredImg3)
plt.title('Box Filtered Image')

# save images
cv2.imwrite('Horizontal_Convolution.png', filteredImg1*255)
cv2.imwrite('Vertical_Convolution.png', filteredImg2*255)
cv2.imwrite('Box_Convolution.png', filteredImg3*255)
```
