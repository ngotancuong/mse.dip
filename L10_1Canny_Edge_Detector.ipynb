```python
import cv2
import numpy as np
import urllib.request
import os

# download the image from the URL
url = 'https://raw.githubusercontent.com/ngotancuong/mse.dip/main/images/bike.png'
filename = 'bike.png'
if not os.path.exists(filename):
    urllib.request.urlretrieve(url, filename)

# load image
Img = cv2.imread(filename)

# loop over some different standard deviation of Gaussian
sigmaArray = [np.sqrt(2), np.sqrt(16), np.sqrt(32)]
threshold = 0.15

# create a figure with subplots
fig, axs = plt.subplots(2, 2)

# show the original image in the first subplot
axs[0, 0].imshow(cv2.cvtColor(Img, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Original Image')

# loop through different sigma values and show the Canny edges in the rest of the subplots
for i, sigma in enumerate(sigmaArray):
    # compute Canny edges
    bw = cv2.Canny(Img, threshold, sigma)
    # show the Canny edges in the subplot
    axs[(i+1)//2, (i+1)%2].imshow(bw, cmap='gray')
    axs[(i+1)//2, (i+1)%2].set_title('sigma = {:.2f}'.format(sigma))
    # save the Canny edges as an image
    cv2.imwrite('Canny_Edge_{}.jpg'.format(sigma), bw)

# show the figure
plt.show()
```
