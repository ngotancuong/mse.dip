```python
import matplotlib.pyplot as plt
import cv2
import urllib.request
import numpy as np

#load image
url = 'https://raw.githubusercontent.com/ngotancuong/mse.dip/main/images/traces.png'
with urllib.request.urlopen(url) as url:
    s = url.read()
    img_array = np.asarray(bytearray(s), dtype=np.uint8)
Img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

#perform erosion with different structuring element (SE)
#a. Small square with edge = 5
se1 = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
BW1 = cv2.erode(Img, se1)
#b. Large square with edge = 10
se2 = cv2.getStructuringElement(cv2.MORPH_RECT, (10, 10))
BW2 = cv2.erode(Img, se2)

#show and save images
fig, ax = plt.subplots(1, 3, figsize=(12, 8))
ax[0].imshow(Img)
ax[0].set_title('Original Image')
ax[1].imshow(BW1)
ax[1].set_title('Erosion by Small Square')
ax[2].imshow(BW2)
ax[2].set_title('Erosion by Large Square')
for i in range(3):
    ax[i].axis('off')
plt.savefig('erosion.png', dpi=300)
plt.show()

#save images
cv2.imwrite('Erosion_Square_5.png', BW1)
cv2.imwrite('Erosion_Square_10.png', BW2)
```
